```python
import warnings

import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import nsfg
import mystats

warnings.simplefilter(action="ignore", category=FutureWarning)
```

```python
var_info = [
    ("time", 1, 8, int),
    ("sex", 9, 16, int),
    ("weight_g", 17, 24, int),
    ("minutes", 25, 32, int),
]
columns = ["name", "start", "end", "type"]
variables = pd.DataFrame(var_info, columns=columns)
variables
```

```python
variables.end += 1
variables
```

```python
dct = mystats.FixedWidthVariables(variables, index_base=1)
dct
```

```python
df = dct.ReadFixedWidth("data/babyboom.dat", skiprows=59)
df
```

```python
diffs = df.minutes.diff()
diffs
```

```python
d = {}
for x in diffs:
    d[x] = d.get(x,0) + 1

total = sum(d.values())
for key, value in d.items():
    d[key] = value / total
d
```

```python
sorted(d.items())
```

```python
xs, freq = zip(*sorted(d.items()))
xs, freq
```

```python
import numpy as np
from numpy.core.fromnumeric import cumsum
ps = cumsum(freq, dtype=np.float)
ps
```

```python
import matplotlib.pyplot as plt
plt.plot(xs, ps, label="babecome", alpha=0.5)
```
